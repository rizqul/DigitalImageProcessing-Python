```python
import matplotlib.pyplot as plt
import imageio
import scipy.ndimage as nd
import numpy as np
%matplotlib inline
```

```python
import imageio
import numpy as np
im = imageio.imread('bird.jpeg').astype(np.float64) / 255.0
```

```python
import matplotlib.pyplot as pyplot
r = im[:, :, 0]
pyplot.imshow(r, cmap='gray')
```

```python
g = im[:, :, 1] + 0.1
pyplot.imshow(g, cmap='gray')
```

```python
temp = r/g
pyplot.imshow(temp, cmap='gray')
```

```python
#mengubah setiap unsur yg < 1 menjadi false dan yg > 1 jadi true
#dikalikan dengan temp agar mengubah 0 menjadi 1 dan 1 menjadi 0

r = (temp > 1) * temp
r = r / r.max() # normalize
print(r)
plt.imshow(r, cmap='gray')
```

```python
#konversi citra ke gambar gray scale

gray = np.dot(img[:, :, :], [0.2126, 0.7512, 0.0722])
plt.imshow(gray, cmap='gray')
```

```python
#mengubah unsur gray kedalam integer 8 bit
#sehingga memperjelas batasan2 unsur gambar

gray_int8 = (gray * 255).astype(np.uint8) & 0xc0
gray_int8 = gray_int8.astype(np.float64) / 255.0
plt.imshow(gray_int8, cmap='gray')
```

```python
#fungsi 2
#mengambil semua unsur yg jumlahnya == 0 atau warna hitam (objek burung dalam gambar)

brd = gray_int8 == 0
plt.imshow(brd, cmap='gray')
```

```python
#fungsi 3

bg = (1 - brd)*gray_int8
plt.imshow(bg, cmap='gray')
```
